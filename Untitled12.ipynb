```python
# ================================
# COMPLETE DEPENDENCY GRAPH BUILDER - SINGLE CELL VERSION
# Run this entire cell in Google Colab
# ================================

# Install dependencies
print("üì¶ Installing dependencies...")
import subprocess
import sys

subprocess.check_call([sys.executable, "-m", "pip", "install", "-q",
                      "tree-sitter-languages", "networkx", "matplotlib",
                      "plotly", "pandas", "pyvis"])
print("‚úÖ Dependencies installed!\n")

# Import all required libraries
from tree_sitter_languages import get_language, get_parser
import networkx as nx
from pathlib import Path
from typing import Dict, List, Set, Tuple, Optional
from collections import defaultdict
import json
import matplotlib.pyplot as plt
import pandas as pd

try:
    from pyvis.network import Network
    HAS_PYVIS = True
except:
    HAS_PYVIS = False

# ================================
# CORE CLASSES
# ================================

class FileDependencyGraphBuilder:
    """
    Build file-level dependency graphs from multi-language projects
    Supports: Python, Java, JavaScript, TypeScript, Go, Rust, C++, C, C#, Ruby
    """

    LANG_CONFIG = {
        'python': {
            'extensions': ['.py'],
            'ts_name': 'python',
            'import_types': ['import_statement', 'import_from_statement'],
            'function_types': ['function_definition'],
            'call_types': ['call'],
        },
        'java': {
            'extensions': ['.java'],
            'ts_name': 'java',
            'import_types': ['import_declaration'],
            'function_types': ['method_declaration'],
            'call_types': ['method_invocation'],
        },
        'javascript': {
            'extensions': ['.js', '.jsx'],
            'ts_name': 'javascript',
            'import_types': ['import_statement'],
            'function_types': ['function_declaration', 'method_definition'],
            'call_types': ['call_expression'],
        },
        'typescript': {
            'extensions': ['.ts', '.tsx'],
            'ts_name': 'typescript',
            'import_types': ['import_statement'],
            'function_types': ['function_declaration', 'method_definition'],
            'call_types': ['call_expression'],
        },
        'go': {
            'extensions': ['.go'],
            'ts_name': 'go',
            'import_types': ['import_declaration'],
            'function_types': ['function_declaration'],
            'call_types': ['call_expression'],
        },
        'rust': {
            'extensions': ['.rs'],
            'ts_name': 'rust',
            'import_types': ['use_declaration'],
            'function_types': ['function_item'],
            'call_types': ['call_expression'],
        },
        'cpp': {
            'extensions': ['.cpp', '.cc', '.cxx', '.hpp', '.h', '.hxx'],
            'ts_name': 'cpp',
            'import_types': ['preproc_include'],
            'function_types': ['function_definition'],
            'call_types': ['call_expression'],
        },
        'c': {
            'extensions': ['.c', '.h'],
            'ts_name': 'c',
            'import_types': ['preproc_include'],
            'function_types': ['function_definition'],
            'call_types': ['call_expression'],
        },
        'c_sharp': {
            'extensions': ['.cs'],
            'ts_name': 'c_sharp',
            'import_types': ['using_directive'],
            'function_types': ['method_declaration'],
            'call_types': ['invocation_expression'],
        },
        'ruby': {
            'extensions': ['.rb'],
            'ts_name': 'ruby',
            'import_types': ['call'],
            'function_types': ['method'],
            'call_types': ['call'],
        },
    }

    def __init__(self):
        """Initialize with pre-built language parsers"""
        self.parsers = {}

        print("üîß Loading language parsers...")

        for lang_name, config in self.LANG_CONFIG.items():
            try:
                ts_name = config['ts_name']
                parser = get_parser(ts_name)
                self.parsers[lang_name] = parser
                print(f"  ‚úì {lang_name}")
            except Exception as e:
                pass

        self.graph = nx.DiGraph()
        self.file_functions = defaultdict(set)
        self.function_to_file = {}
        self.file_imports = defaultdict(set)

        print(f"\n‚úÖ Initialized with {len(self.parsers)} language parsers\n")

    def detect_language(self, filepath: Path) -> Optional[str]:
        """Detect language from file extension"""
        ext = filepath.suffix.lower()
        for lang, config in self.LANG_CONFIG.items():
            if ext in config['extensions']:
                return lang
        return None

    def parse_file(self, filepath: Path) -> Tuple[Optional[any], Optional[bytes], Optional[str]]:
        """Parse a source file and return AST"""
        language = self.detect_language(filepath)

        if not language or language not in self.parsers:
            return None, None, None

        try:
            with open(filepath, 'rb') as f:
                source = f.read()

            parser = self.parsers[language]
            tree = parser.parse(source)
            return tree.root_node, source, language

        except Exception as e:
            return None, None, None

    def extract_functions(self, node, source: bytes, language: str) -> Set[str]:
        """Extract all function names from AST"""
        functions = set()
        config = self.LANG_CONFIG.get(language, {})
        function_types = config.get('function_types', [])

        def traverse(n):
            if n.type in function_types:
                name = self._get_function_name(n, source, language)
                if name:
                    functions.add(name)
            for child in n.children:
                traverse(child)

        traverse(node)
        return functions

    def _get_function_name(self, node, source: bytes, language: str) -> Optional[str]:
        """Extract function name from function node"""
        name_node = node.child_by_field_name('name')

        if not name_node:
            for child in node.children:
                if child.type == 'identifier':
                    name_node = child
                    break

        if name_node:
            try:
                return source[name_node.start_byte:name_node.end_byte].decode('utf-8', errors='ignore')
            except:
                return None
        return None

    def extract_function_calls(self, node, source: bytes, language: str) -> Set[str]:
        """Extract all function call names from AST"""
        calls = set()
        config = self.LANG_CONFIG.get(language, {})
        call_types = config.get('call_types', [])

        def traverse(n):
            if n.type in call_types:
                call_name = self._get_call_name(n, source, language)
                if call_name:
                    calls.add(call_name)
            for child in n.children:
                traverse(child)

        traverse(node)
        return calls

    def _get_call_name(self, node, source: bytes, language: str) -> Optional[str]:
        """Extract function name from call expression"""
        func_node = node.child_by_field_name('function')

        if not func_node:
            func_node = node.children[0] if node.children else None

        if not func_node:
            return None

        try:
            call_text = source[func_node.start_byte:func_node.end_byte].decode('utf-8', errors='ignore')
            if '.' in call_text:
                call_text = call_text.split('.')[-1]
            return call_text
        except:
            return None

    def build_from_directory(self, repo_path: str,
                            exclude_dirs: Optional[Set[str]] = None) -> nx.DiGraph:
        """Build file dependency graph from directory"""

        if exclude_dirs is None:
            exclude_dirs = {
                'node_modules', '.git', '__pycache__', 'venv', 'env',
                'build', 'dist', '.pytest_cache', '.venv', 'target',
                'out', 'bin', 'obj', '.idea', '.vscode', 'vendor',
                '.next', 'coverage', '.gradle', '.mvn'
            }

        repo_path = Path(repo_path)

        if not repo_path.exists():
            print(f"‚ùå Path does not exist: {repo_path}")
            return self.graph

        print(f"\n{'='*60}")
        print(f"üìÇ Scanning: {repo_path}")
        print(f"{'='*60}")

        # Collect all source files
        all_files = []
        for config in self.LANG_CONFIG.values():
            for ext in config['extensions']:
                all_files.extend(repo_path.rglob(f'*{ext}'))

        # Filter excluded directories
        source_files = [
            f for f in all_files
            if not any(excluded in f.parts for excluded in exclude_dirs)
        ]

        print(f"üìÑ Found {len(source_files)} source files")

        if len(source_files) == 0:
            print("‚ö†Ô∏è  No source files found!")
            return self.graph

        # Show language breakdown
        lang_counts = defaultdict(int)
        for f in source_files:
            lang = self.detect_language(f)
            if lang:
                lang_counts[lang] += 1

        print(f"\nüìä Language breakdown:")
        for lang, count in sorted(lang_counts.items(), key=lambda x: x[1], reverse=True):
            print(f"  ‚Ä¢ {lang}: {count} files")

        # Phase 1: Extract functions
        print(f"\n{'='*60}")
        print("Phase 1: Extracting functions")
        print(f"{'='*60}")

        processed = 0
        for filepath in source_files:
            try:
                self._process_file(filepath)
                processed += 1
                if processed % 100 == 0:
                    print(f"  Processed {processed}/{len(source_files)} files...")
            except:
                pass

        print(f"  ‚úì Processed: {processed} files")
        print(f"  ‚úì Total functions: {len(self.function_to_file)}")

        # Phase 2: Build dependencies
        print(f"\n{'='*60}")
        print("Phase 2: Building dependencies")
        print(f"{'='*60}")

        dep_count = 0
        for filepath in source_files:
            try:
                deps = self._build_file_dependencies(filepath)
                dep_count += len(deps)
            except:
                pass

        print(f"  ‚úì Created {dep_count} dependency connections")

        # Print summary
        print(f"\n{'='*60}")
        print("GRAPH SUMMARY")
        print(f"{'='*60}")
        print(f"  üìä Files: {self.graph.number_of_nodes()}")
        print(f"  üîó Dependencies: {self.graph.number_of_edges()}")

        if self.graph.number_of_nodes() > 0:
            avg_deps = self.graph.number_of_edges() / self.graph.number_of_nodes()
            print(f"  üìà Avg dependencies per file: {avg_deps:.1f}")

        return self.graph

    def _process_file(self, filepath: Path):
        """Process a single file to extract functions"""
        root_node, source, language = self.parse_file(filepath)

        if not root_node:
            return

        filepath_str = str(filepath)

        functions = self.extract_functions(root_node, source, language)
        self.file_functions[filepath_str] = functions

        for func_name in functions:
            self.function_to_file[func_name] = filepath_str

        if not self.graph.has_node(filepath_str):
            self.graph.add_node(
                filepath_str,
                filename=filepath.name,
                language=language,
                num_functions=len(functions)
            )

    def _build_file_dependencies(self, filepath: Path) -> List[Tuple[str, str]]:
        """Build dependencies for a single file"""
        root_node, source, language = self.parse_file(filepath)

        if not root_node:
            return []

        filepath_str = str(filepath)
        dependencies = []

        calls = self.extract_function_calls(root_node, source, language)

        for call_name in calls:
            if call_name in self.function_to_file:
                target_file = self.function_to_file[call_name]

                if target_file != filepath_str:
                    if self.graph.has_edge(filepath_str, target_file):
                        self.graph[filepath_str][target_file]['weight'] += 1
                    else:
                        self.graph.add_edge(filepath_str, target_file, weight=1)
                        dependencies.append((filepath_str, target_file))

        return dependencies

    def get_file_info(self, filepath: str) -> Optional[Dict]:
        """Get detailed information about a file"""
        if filepath not in self.graph:
            return None

        return {
            'file': filepath,
            'filename': Path(filepath).name,
            'metadata': self.graph.nodes[filepath],
            'depends_on': list(self.graph.successors(filepath)),
            'imported_by': list(self.graph.predecessors(filepath)),
            'in_degree': self.graph.in_degree(filepath),
            'out_degree': self.graph.out_degree(filepath),
        }

    def export_to_json(self, output_path: str):
        """Export graph to JSON"""
        data = {
            'graph': nx.node_link_data(self.graph),
            'statistics': {
                'total_files': self.graph.number_of_nodes(),
                'total_dependencies': self.graph.number_of_edges(),
                'total_functions': len(self.function_to_file),
            }
        }

        with open(output_path, 'w') as f:
            json.dump(data, f, indent=2)

        print(f"‚úÖ Exported to {output_path}")

    def get_most_connected_files(self, top_n: int = 10) -> List[Tuple[str, int]]:
        """Get files with most dependencies"""
        degrees = [(node, self.graph.degree(node)) for node in self.graph.nodes()]
        return sorted(degrees, key=lambda x: x[1], reverse=True)[:top_n]

    def get_most_depended_on_files(self, top_n: int = 10) -> List[Tuple[str, int]]:
        """Get files that are most depended upon"""
        in_degrees = [(node, self.graph.in_degree(node)) for node in self.graph.nodes()]
        return sorted(in_degrees, key=lambda x: x[1], reverse=True)[:top_n]


class DependencyVisualizer:
    """Visualize file dependency graphs"""

    @staticmethod
    def plot_graph_matplotlib(graph: nx.DiGraph, title: str = "File Dependency Graph"):
        """Plot using matplotlib (good for small graphs < 50 nodes)"""
        if graph.number_of_nodes() == 0:
            print("‚ùå Graph is empty")
            return

        if graph.number_of_nodes() > 50:
            print("‚ö†Ô∏è  Graph too large for matplotlib")
            return

        plt.figure(figsize=(20, 15))
        pos = nx.spring_layout(graph, k=2, iterations=50, seed=42)
        in_degrees = [graph.in_degree(node) for node in graph.nodes()]

        nx.draw_networkx_nodes(graph, pos, node_color=in_degrees, node_size=500,
                              cmap='YlOrRd', alpha=0.8)
        nx.draw_networkx_edges(graph, pos, edge_color='gray', alpha=0.3,
                              arrows=True, arrowsize=10, width=0.5)

        labels = {node: Path(node).name for node in graph.nodes()}
        nx.draw_networkx_labels(graph, pos, labels, font_size=6)

        plt.title(title, fontsize=16)
        plt.axis('off')
        plt.tight_layout()
        plt.show()

    @staticmethod
    def plot_interactive(graph: nx.DiGraph, output_file: str = '/content/dependency_graph.html'):
        """Create interactive HTML visualization"""
        if not HAS_PYVIS:
            print("‚ùå PyVis not installed")
            return None

        if graph.number_of_nodes() == 0:
            print("‚ùå Graph is empty")
            return None

        print(f"üé® Creating interactive visualization...")

        net = Network(height='800px', width='100%', directed=True,
                     bgcolor='#222222', font_color='white')
        net.barnes_hut(gravity=-8000, central_gravity=0.3, spring_length=250)

        for node in graph.nodes():
            in_deg = graph.in_degree(node)
            out_deg = graph.out_degree(node)

            color = '#ff4444' if in_deg > 5 else '#ffaa00' if in_deg > 2 else '#4444ff'
            size = 10 + (in_deg + out_deg) * 2

            net.add_node(node, label=Path(node).name,
                        title=f"{Path(node).name}\nDeps: {out_deg}\nDependent: {in_deg}",
                        color=color, size=size)

        for source, target in graph.edges():
            weight = graph[source][target].get('weight', 1)
            net.add_edge(source, target, value=weight)

        net.save_graph(output_file)
        print(f"‚úÖ Interactive graph saved to: {output_file}")
        return output_file

    @staticmethod
    def plot_statistics(graph: nx.DiGraph):
        """Plot statistics"""
        if graph.number_of_nodes() == 0:
            print("‚ùå Graph is empty")
            return

        in_degrees = [graph.in_degree(node) for node in graph.nodes()]
        out_degrees = [graph.out_degree(node) for node in graph.nodes()]

        fig, axes = plt.subplots(2, 2, figsize=(15, 10))

        axes[0, 0].hist(in_degrees, bins=20, color='skyblue', edgecolor='black')
        axes[0, 0].set_title('Files by Dependents')
        axes[0, 0].set_xlabel('Dependents')
        axes[0, 0].set_ylabel('Count')

        axes[0, 1].hist(out_degrees, bins=20, color='lightcoral', edgecolor='black')
        axes[0, 1].set_title('Files by Dependencies')
        axes[0, 1].set_xlabel('Dependencies')
        axes[0, 1].set_ylabel('Count')

        in_deg_sorted = sorted([(node, graph.in_degree(node)) for node in graph.nodes()],
                               key=lambda x: x[1], reverse=True)[:10]

        if in_deg_sorted:
            files, degs = zip(*in_deg_sorted)
            files = [Path(f).name for f in files]
            axes[1, 0].barh(files, degs, color='green', alpha=0.7)
            axes[1, 0].set_title('Top 10 Most Depended-On')
            axes[1, 0].set_xlabel('Dependents')
            axes[1, 0].invert_yaxis()

        languages = {}
        for node in graph.nodes():
            lang = graph.nodes[node].get('language', 'unknown')
            languages[lang] = languages.get(lang, 0) + 1

        if languages:
            axes[1, 1].pie(languages.values(), labels=languages.keys(),
                          autopct='%1.1f%%', startangle=90)
            axes[1, 1].set_title('Language Distribution')

        plt.tight_layout()
        plt.show()



def analyze_project(graph: nx.DiGraph, builder: FileDependencyGraphBuilder):
    """Comprehensive project analysis"""
    if graph.number_of_nodes() == 0:
        print("‚ùå No files in graph")
        return

    print(f"\n{'='*60}")
    print("PROJECT ANALYSIS")
    print(f"{'='*60}")

    print(f"\nüìä Overall Statistics:")
    print(f"  ‚Ä¢ Total files: {graph.number_of_nodes()}")
    print(f"  ‚Ä¢ Total dependencies: {graph.number_of_edges()}")
    print(f"  ‚Ä¢ Total functions: {len(builder.function_to_file)}")

    if graph.number_of_nodes() > 0:
        density = nx.density(graph)
        print(f"  ‚Ä¢ Graph density: {density:.3f}")

    print(f"\nüî§ Language Breakdown:")
    languages = {}
    for node in graph.nodes():
        lang = graph.nodes[node].get('language', 'unknown')
        languages[lang] = languages.get(lang, 0) + 1

    for lang, count in sorted(languages.items(), key=lambda x: x[1], reverse=True):
        pct = (count / graph.number_of_nodes()) * 100
        print(f"  ‚Ä¢ {lang}: {count} files ({pct:.1f}%)")

    print(f"\nüîó Most Connected Files:")
    top_files = builder.get_most_connected_files(5)
    for filepath, degree in top_files:
        print(f"  ‚Ä¢ {Path(filepath).name}: {degree} connections")

    print(f"\n‚≠ê Most Depended-On Files:")
    top_deps = builder.get_most_depended_on_files(5)
    for filepath, in_degree in top_deps:
        if in_degree > 0:
            print(f"  ‚Ä¢ {Path(filepath).name}: {in_degree} dependents")


def export_analysis(graph: nx.DiGraph, builder: FileDependencyGraphBuilder):
    """Export analysis results"""
    print(f"\n{'='*60}")
    print("EXPORTING RESULTS")
    print(f"{'='*60}")

    builder.export_to_json('/content/dependency_graph.json')

    file_data = []
    for node in graph.nodes():
        file_data.append({
            'file': Path(node).name,
            'full_path': node,
            'language': graph.nodes[node].get('language', 'unknown'),
            'num_functions': graph.nodes[node].get('num_functions', 0),
            'dependencies': graph.out_degree(node),
            'dependents': graph.in_degree(node),
            'total_connections': graph.degree(node)
        })

    df = pd.DataFrame(file_data)
    csv_path = '/content/file_dependencies.csv'
    df.to_csv(csv_path, index=False)
    print(f"‚úÖ Exported CSV to: {csv_path}")

    print(f"\nüìÑ Top 10 by connections:")
    print(df.sort_values('total_connections', ascending=False).head(10).to_string())


def analyze_custom_project(project_path: str):
    """
    Analyze any project folder

    Usage:
    analyze_custom_project('/content/your_project_folder')
    """
    print("\n" + "="*60)
    print(f"ANALYZING PROJECT: {project_path}")
    print("="*60)

    builder = FileDependencyGraphBuilder()
    graph = builder.build_from_directory(project_path)

    if graph.number_of_nodes() == 0:
        print("\n‚ùå No source files found!")
        return None, None

    analyze_project(graph, builder)
    export_analysis(graph, builder)

    visualizer = DependencyVisualizer()
    visualizer.plot_interactive(graph)

    if graph.number_of_nodes() <= 50:
        visualizer.plot_graph_matplotlib(graph, f"Dependencies: {Path(project_path).name}")

    visualizer.plot_statistics(graph)

    return builder, graph


def query_file_dependencies(builder: FileDependencyGraphBuilder, graph: nx.DiGraph, filename: str):
    """Query dependencies for a specific file"""
    matching_files = [node for node in graph.nodes() if filename in node]

    if not matching_files:
        print(f"‚ùå File '{filename}' not found")
        return

    if len(matching_files) > 1:
        print(f"‚ö†Ô∏è  Multiple files match '{filename}':")
        for f in matching_files:
            print(f"  ‚Ä¢ {f}")
        return

    filepath = matching_files[0]
    info = builder.get_file_info(filepath)

    print(f"\n{'='*60}")
    print(f"FILE ANALYSIS: {Path(filepath).name}")
    print(f"{'='*60}")

    print(f"\nüìç Location: {filepath}")
    print(f"üî§ Language: {info['metadata'].get('language', 'unknown')}")
    print(f"üìä Functions: {info['metadata'].get('num_functions', 0)}")
    print(f"üì• Depends on: {info['out_degree']} files")
    print(f"üì§ Depended by: {info['in_degree']} files")

    if info['depends_on']:
        print(f"\n‚¨áÔ∏è  This file depends on:")
        for dep in info['depends_on'][:10]:
            print(f"  ‚Ä¢ {Path(dep).name}")

    if info['imported_by']:
        print(f"\n‚¨ÜÔ∏è  This file is imported by:")
        for imp in info['imported_by'][:10]:
            print(f"  ‚Ä¢ {Path(imp).name}")
```
